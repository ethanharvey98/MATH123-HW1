(d)

```python
def truncated_svd(u, s, vh, n_components):
    new_u = u[:,0:n_components]
    new_s = np.diag(s)[0:n_components,0:n_components]
    new_vh = vh[0:n_components,:]
    return np.dot(np.dot(new_u,new_s),new_vh)
    
plt.imshow(truncated_svd(u, s, vh, 10), cmap='gray')
# Display the plot
plt.show()
```

The image quality is not great. The panises are barely visible. The image is grainy and pixelated.

The image quality when k = 50 is significantly better than when k = 25. The panises are totally recognizable and the image is not grainy or pixelated.

Assuming that storing a single pixel costs 1 byte of memory, we would need __m x n bytes__ to store a general m x n matrix __A__. We would need __(m x r) + (r<sup>2</sup>) + (r x n) bytes__ to store the reduced SVD of __A__. We would need __(m x k) + (k<sup>2</sup>) + (k x n) bytes__ to store the truncated SVD of __A__. We would need __1,871,800 bytes__ to store the panises image matrix. We would need __4,873,050 bytes__ to store the reduced SVD of the panises image. We would need __915,900 bytes__ to store the truncated SVD of the panises image with a k = 300. If we chose a smaller k our truncated SVD could be even smaller. The computatinal advantages of working with a truncated SVD include:
* A truncated SVD takes up less storage than a general m x n matrix
* Using a truncated SVD as features for a model significantly decreases the amount of features and therefore training time
